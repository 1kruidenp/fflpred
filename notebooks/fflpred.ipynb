# ARMIN !!!

```python
raw_teams=[players_raw_16, players_raw_17_18, players_raw_18_19, players_raw_19_20, players_raw_20_21]

for teams in raw_teams: #Iterate through the different raw files
    teams['name']=teams['first_name']+' '+teams['second_name'] #Merge the first and second names of the raw players
    teams['name']=teams['name'].str.lower() #Set these names to lower
    for i,row in teams.iterrows(): #Change caglar to çaglar
        if row['name']=='caglar söyüncü':
            teams['name'].at[i]='çaglar söyüncü'

#Here we split the data frame into the seasons and join the raw positions on the name

df17=full_data.loc[full_data['season'] == 17].copy()

df18=full_data.loc[full_data['season'] == 18].copy()

df19=full_data.loc[full_data['season'] == 19].copy()

df20=full_data.loc[full_data['season'] == 20].copy()
abc=players_raw_19_20[['name','team']]
abc.set_index('name',drop=True,inplace=True)
df20=df20.join(abc, on=['name'])
df20.rename(columns={'team':'team_id'}, inplace=True)

xyz=team_19_20[['id','name']].copy()
xyz.rename(columns={'name':'team_name','id':'team_id'}, inplace=True)
xyz.set_index('team_id', drop=True, inplace=True)
df20=df20.join(xyz, on=['team_id'])

df21=full_data.loc[full_data['season'] == 21].copy()
abc=players_raw_20_21[['name','team']]
abc.set_index('name',drop=True,inplace=True)
df21=df21.join(abc, on=['name'])
df21.rename(columns={'team':'team_id'}, inplace=True)

xyz_2=team_20_21[['id','name']].copy()
xyz_2.rename(columns={'name':'team_name','id':'team_id'}, inplace=True)
xyz_2.set_index('team_id', drop=True, inplace=True)
df21=df21.join(xyz_2, on=['team_id'])

#Reconcatenate the data into a complete dataset

complete_data_teams = pd.concat([df17,df18,df19,df20,df21])
complete_data_teams.drop_duplicates(inplace=True) #Drop the duplicates (Danny Ward and Ben Davies)
```
