```python
features = ['name', 'assists', 'clean_sheets', 'goals_scored',
            'goals_conceded', 'penalties_missed', 'penalties_saved',
            'value', 'opponent_level', 'total_points']
```

```python
data_features = data[features]
data_features.head()
```

```python
mean_data = data_features.groupby('name').rolling(5, closed = 'left').mean()
mean_data
```

```python
df = mean_data.reset_index()
df
```

```python
df.head(10)
```

```python
result = map(lambda gw:gw, data['GW'])
df['GW'] = list(result)

result_1 = map(lambda season:season, data['season'])
df['season'] = list(result_1)

result_2 = map(lambda home:home, data['was_home'])
df['was_home'] = list(result_2)

result_3 = map(lambda position:position, data['position'])
df['position'] = list(result_3)
```

```python
df.head(10)
```
